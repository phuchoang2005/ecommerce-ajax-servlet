```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root123",
)
conn.cursor().execute("use ecommerce_new")
```

```python
import pandas as pd
def execute_query(title, sql):
    print(f"🔍 {title}")
    try:
            # Sử dụng pandas để đọc và hiển thị cho đẹp
        df = pd.read_sql(sql, conn)
        if df.empty:
            print("Không tìm thấy kết quả phù hợp.")
        else:
            display(df) # Hàm display chuyên dùng cho Jupyter
        print("-" * 50)
    except Exception as e:
        print(f"❌ Lỗi: {e}")
```

```python
execute_query("Select all users", "select * from users;")
```

```python
execute_query("SELECT ALL PROFILES", "select * from profiles;")
```
